code for gettting the relative humidity data

```python
OUTPUT_FILE_NAME = "../data/relative-humidity.csv"

for file in json_files:
    print(f"processing {file}")
    with open(os.path.join(folder_path, file), "r") as file:
        data = json.load(file)

        if data is None:
            print(f"Error reading file {file}")
            continue
    
        for reading in data['data']['readings']:
            current_timestamp_readings_json = []
            timestamp = reading['timestamp']
            print(timestamp)

            for station in reading['data']:
                station_id = station['stationId']
                value = station['value']

                current_timestamp_readings_json.append({'StationId': station_id, 'Value': value, 'Timestamp': timestamp})
            
            rh_df = pd.DataFrame(current_timestamp_readings_json)
            if not os.path.exists(OUTPUT_FILE_NAME):
                rh_df.to_csv(OUTPUT_FILE_NAME, index=False)
            else:
                rh_df.to_csv(OUTPUT_FILE_NAME, index=False, mode='a', header=False)
```
